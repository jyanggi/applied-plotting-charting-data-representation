```python
import pandas as pd
import numpy as np
pd.set_option('display.float_format', lambda x: '%.5f' % x)
gdp = pd.read_csv('worldbank_gdp.csv')
gdp.fillna(0, inplace = True)
gdp = gdp.loc[gdp['Country Name']=='United States' ]
gdp.drop(['Country Name','Country Code', 'Indicator Name', 'Indicator Code'] +  [f'{i}' for i in range(1960, 2022) if i < 1990] , axis=1, inplace = True)
gdp = gdp.loc[:, ~gdp.columns.str.contains('^Unnamed')]
gdp = gdp.T.reset_index(0)
gdp.rename(columns={"index": "year", 251: "gdp"}, errors="raise", inplace=True)
gdp.index = gdp.year
gdp.drop('year' , axis=1, inplace = True)
print(gdp)
```

```python
inflation = pd.read_csv('worldbank_inflation.csv')
inflation.fillna(0, inplace = True)
inflation = inflation[inflation['Country Name']=='United States']
inflation.drop(['Country Name','Country Code', 'Indicator Name', 'Indicator Code'] +  [f'{i}' for i in range(1960, 2022) if i < 1990] , axis=1, inplace = True)
inflation = inflation.loc[:, ~inflation.columns.str.contains('^Unnamed')]
inflation = inflation.T.reset_index(0)
inflation.rename(columns={"index": "year", 251: "inflation rate"}, errors="raise", inplace=True)
inflation.index = inflation.year
inflation.drop('year' , axis=1, inplace = True)
print(inflation)
```

# United States, GDP and inflation rate

What is the correlation between the GDP and inflation rate of  the United States in the year 1990 onwards?

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from matplotlib.backends.backend_agg import FigureCanvasAgg


plt.figure(figsize=(20,15))
plt.title('United States GDP and Inflation rate (1990-2021)', fontsize=25)
data = inflation.join(gdp, how='outer')
data = data.interpolate(method='linear')
print(data)
plt.xticks(range(0,len(data.index)), data.index,  fontsize=14)
ax = data['inflation rate'].plot(c='red')
ax.set_xlabel('Year' ,labelpad=50, fontsize=18)
ax.set_ylabel('Inflation Rate',labelpad=50, fontsize=18)
ax.set_facecolor('xkcd:white')


ax2 = data['gdp'].plot(ax=ax, secondary_y=True, c='blue')
ax2.set_ylabel('GDP',labelpad=50, fontsize=18)
ax2.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(float(x), ',')))

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
handles, labels = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax.legend(handles=(handles+handles2), labels=(labels+labels2), loc='upper left')
corr = data.corr()['inflation rate']['gdp'].round(decimals=5)    
corr_label = f'Correlation: {corr}'
print(corr)
ax.text(1,1,corr_label,fontsize=15)

x = ax.xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
#canvas = FigureCanvasAgg(f)
#canvas.print_png('gdp_inflation.png')
plt.show()
```

# Conclusion


Based on the dataset from 1990 to 2021 -0.34348 indicates a weak negative correlation between the gdp and inflation rate. This suggests that as the inflation rate increases, the GDP tends to decrease and vice versa, but the relationship is not that strong indicating that there may be other factors that affect the relationship between inflation rate and GDP.
